Question 1: **Book Class**

```python
class Book:
    def __init__(self, title, author, isbn, availability=True):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.__availability = availability

    def get_availability(self):
        return self.__availability

    def set_availability(self, availability):
        self.__availability = availability

    def display_info(self):
        print("Title:", self.title)
        print("Author:", self.author)
        print("ISBN:", self.isbn)
        print("Availability:", "Available" if self.__availability else "Not Available")
        print()


# Creating an instance of the Book class
book1 = Book("Python Programming", "John Doe", "1234567890")

# Accessing availability using getter method
print("Availability:", book1.get_availability())

# Updating availability using setter method
book1.set_availability(False)
print("Updated Availability:", book1.get_availability())

# Displaying book information
book1.display_info()

```

Question 2: **Library class**

```python
class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)
        print("Book added successfully!")

    def remove_book(self, isbn):
        for book in self.books:
            if book.isbn == isbn:
                self.books.remove(book)
                print("Book removed successfully!")
                return
        print("Book with ISBN {} not found.".format(isbn))

    def list_books(self):
        if not self.books:
            print("No books in the library.")
            return
        print("Books in the library:")
        for book in self.books:
            book.display_info()

    def search_book(self, keyword):
        found_books = []
        for book in self.books:
            if keyword.lower() in book.title.lower() or keyword.lower() in book.author.lower():
                found_books.append(book)
        if not found_books:
            print("No books found with the keyword '{}'.".format(keyword))
            return
        print("Books found with the keyword '{}':".format(keyword))
        for book in found_books:
            book.display_info()

# Creating instance of Library class
library = Library()

# Adding books to the library
library.add_book(Book("Python Programming", "Joe Dassin", "1234567890"))
library.add_book(Book("Java Programming", "Joy Nkwihoreze", "0987654321"))
library.add_book(Book("Machine Learning", "Teta Ella", "1357924680"))

# Listing all books in the library
library.list_books()

# Searching for a book by title or author
library.search_book("python")
library.search_book("Joy Nkwihoreze")

```

```python
class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)
        print("Book added successfully!")

    def remove_book(self, isbn):
        for book in self.books:
            if book.isbn == isbn:
                self.books.remove(book)
                print("Book removed successfully!")
                return
        print("Book with ISBN {} not found.".format(isbn))

    def list_books(self):
        if not self.books:
            print("No books in the library.")
            return
        print("Books in the library:")
        for book in self.books:
            book.display_info()

    def search_book(self, keyword):
        found_books = []
        for book in self.books:
            if keyword.lower() in book.title.lower() or keyword.lower() in book.author.lower():
                found_books.append(book)
        if not found_books:
            print("No books found with the keyword '{}'.".format(keyword))
            return
        print("Books found with the keyword '{}':".format(keyword))
        for book in found_books:
            book.display_info()

    def check_out_book(self, isbn):
        for book in self.books:
            if book.isbn == isbn:
                if book.get_availability():
                    book.set_availability(False)
                    print("Book with ISBN {} checked out successfully!".format(isbn))
                else:
                    print("Book with ISBN {} is not available for checkout.".format(isbn))
                return
        print("Book with ISBN {} not found.".format(isbn))

    def return_book(self, isbn):
        for book in self.books:
            if book.isbn == isbn:
                book.set_availability(True)
                print("Book with ISBN {} returned successfully!".format(isbn))
                return
        print("Book with ISBN {} not found.".format(isbn))
# Creating instance of Library class
library = Library()

# Adding books to the library
library.add_book(Book("Python Programming", "Joe Dassin", "1234567890"))
library.add_book(Book("Java Programming", "Joy Nkwihoreze", "0987654321"))
library.add_book(Book("Machine Learning", "Teta Ella", "1357924680"))

# Checking out a book
library.check_out_book("1234567890")

# Listing all books in the library
library.list_books()

# Returning a book
library.return_book("1234567890")

# Listing all books in the library after returning the book
library.list_books()

```
